```python
#Delete the columns which is having all rows values as na.
data.dropna(axis=1, how="all")
```

```python
#Finding percentage of null or missing values
null_in_perc = round(100*(data.isnull().sum()/len(data.index)), 2)

#Printing columns which have more than 0% missing values
null_in_perc[ null_in_perc > 0 ]
```

```python
#print columns which has more than 25% null values in it.
null_col_25 = null_in_perc[ null_in_perc > 25 ]
null_col_25
```

```python
#Removing columns which has more than 25% null values in it.
existing_cols = [col for col in null_col_25 if col in data.columns]
data.drop(existing_cols, axis=1, inplace=True)
```
